```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy import stats
```

# OLS on data generated through simulations for
        yi=b0+xb0i+ei
 where x in iidNr(u;suminf)and ei iid N(0, sig^2) for i= 1,.....n.        

```python
sig = 0.5
nsim = 10
Tvec = [10,25,50,100,250]
nvec = [1,2,5]

holding_place_3b = np.zeros((nsim,len(Tvec),len(nvec)))
for ndx,n in enumerate(nvec):
    for tdx,T in enumerate(Tvec):
        for isim in range(nsim):
            sample =  np.random.permutation(2*T)
            train = sample[:int(T)]
            test = sample[int(T):]
            y = np.random.normal(0,sig,size=(2*T,1))
            x = np.random.normal(0,sig,size=(2*T,n))
            x = sm.add_constant(x)
            model = smf.OLS(y[train],x[train]).fit()
            mspe = ((y[test]-model.predict(x[test]))**2).mean()
            holding_place_3b[isim,tdx,ndx] = mspe
print(holding_place_3b.mean(0))
```

```python
# It can be seen that from the above results the mean square error reduces with increase 
# in number of variables
```

# OLS with 2,5 and 10 k-folds on DGP and different bias for eq.
                         yi=b0+xb0i+ei
     where x in iidNr(u;suminf)and ei iid N(0, sig^2) for i= 1,.....n.        

```python
sig = 0.5
nsim = 10
Tvec = [10,25,50,100,250]
kvec = [2,5,10]
nvec = [1,2,5]
bvec = [-1,0,1]
holding_place_3b = np.zeros((nsim,len(Tvec),len(nvec),len(kvec)))
for bdx,c in enumerate(bvec):
    for kdx,k in enumerate(kvec):
        for ndx,n in enumerate(nvec):
            for tdx,T in enumerate(Tvec):
                for isim in range(nsim):
                    mspe = np.zeros(k)
                    for j in range(k):    
                            sample =  np.random.permutation(2*T)
                            train = sample[:int(T)]
                            test = sample[int(T):]
                            train_l = sample[:int((k-1-j)*T/k)]
                            train_u = sample[int((k-j)*T/k):]
                            train = np.concatenate((train_l,train_u),0)
                            test = sample[int((k-1-j)*T/k):int((k-j)*T/k)]
                            sample =  np.random.permutation(T)
                            train = sample[:int(T/2)]
                            test = sample[int(T/2):]
                            x = np.random.normal(0,sig,size=(T,n))
                            b = np.ones((n,1))*bvec[bdx]
                            tg= x.dot(b)
                            y = x.dot(b) + np.random.normal(0,sig,size=(T,1))
                            model = smf.OLS(y[train],x[train]).fit()
                            mspe [j] = ((y[test]-model.predict(x[test]))**2).mean()
                    holding_place_3b[isim,tdx,ndx,kdx]=mspe.mean()
    print("The values below are  mean and standarad deviation for B ",c)
    print(holding_place_3b.mean(0)[:,:,0])
    print(holding_place_3b.mean(0)[:,:,1])
    print(holding_place_3b.mean(0)[:,:,2])
    print()
    print()
    print(holding_place_3b.std(0)[:,:,0])
    print(holding_place_3b.std(0)[:,:,1])
    print(holding_place_3b.std(0)[:,:,2])
```
